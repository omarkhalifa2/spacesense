#### Loading EuroSAT data

1. Built on Sentinel-2 Satellite Imagery from European Space Agency. 

2. Dataset source and reference: https://github.com/phelber/eurosat

3. Same copyright and license applies as described at the dataset source.

```python
import os
os.chdir('../')
```

```python
from spacesense import training_data
```

##### 1.Create an instace of the class

```python
eurosat = training_data.EuroSAT()
```

See all information on the dataset

```python
eurosat.info
```

Two types of dataset is available:
1. All 13 bands of Sentinel-2

2. RGB images of Sentinel-2

##### 2. Download from source

```python
""" files with all 13 sentinel-2 bands"""
eurosat.download_all_bands()  

""" RGB images i.e. files with 3 bands of Red-Green-Blue """
eurosat.download_rgb()
```

##### 3. Load into an X (data) , y (label id) array 

```python
X,y = eurosat.fetch_data(type='all_bands',labels='all') # by default 13_bands and all labels are loaded
```

```python
X_rgb,y_rgb =  eurosat.fetch_data(type='rgb',labels=['Forest','Industrial','River'])
```

###### Load selected labels and rgb data

```python
X_sample,y_sample = eurosat.fetch_data(type='rgb',labels=['Forest','Industrial','River'])
```

##### 4. save data in  .npy format for  faster iterations

```python
eurosat.save_as_npy(X_sample,save_folder='data/',file_name='X_sample')
eurosat.save_as_npy(y_sample,save_folder='data/',file_name='y_sample')
```

##### 5. Display sample images

```python
eurosat.display_sample(labels='all')
```
